# Example usage

Here we will demonstrate how to use `pycounts` to count the words in atext file and plot the top 5 results. 

# Imports

```python
from pycounts_nh.pycounts_nh import count_words
from pycounts_nh.plotting import plot_words
```

# Create a text file

We'll first create a file tow work with using a famous quote from Einstein: 

```python
quote = """ Insanity the same thing over and over and expecting 
    different results."""
with open("einstein.txt", "w") as file: 
    file.write(quote)
```

```python
counts = count_words("einstein.txt")
print(counts)
```

# Plot words

We can now plot the results using `plot_words()` function:

```python
fig = plot_words(counts, n = 5)
```
